```python
data = {
    'Ticker': ['AAPL','MSFT','JPM','TSLA','WMT','GM','RIVN','AXP','GS','COST','TGT','NVDA'],
    'Sector': ['Tech','Tech','Finance','Auto','Retail','Auto','Auto','Finance','Finance','Retail','Retail','Tech'],
    'Market Cap': [3.45,3.77,0.83,1.08,0.77,0.05,0.02,0.23,0.23,0.42,0.04,4.29],
    'YTD Return': [-0.068, 0.1910, 0.2525, -0.1441, 0.0769,0.1005,0.0142,0.1043,0.2865,0.0301,-0.2917,0.2771]
}

# importing pandas and setting the data frame for our market cap graph
import pandas as pd
df = pd.DataFrame(data)
sector_cap = df.groupby('Sector')['Market Cap'].mean()

import matplotlib.pyplot as plt

# creating the Sector cap graph and formatting
ax = sector_cap.plot(kind='bar', title='Avg. Sector Cap (in Trillions)')
ax.set_xlabel('Sector'); ax.set_ylabel('Avg. Market Cap')
bars = plt.bar(sector_cap.index, sector_cap.values)
# Add labels above bars
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,  # x position (center of bar)
        height,                           # y position (top of bar)
        f"{height:.2f}",                  # format number (2 decimals)
        ha='center', va='bottom'          # center horizontally, just above bar
    )
plt.tight_layout(); plt.show()

# Make sure your YTD Return is numeric (in decimals, not % strings)
# Example: df['YTD Return'] = df['YTD Return'].str.replace('%','').astype(float) / 100

# Create boxplot
df.boxplot(column='YTD Return', by='Sector')

plt.title("Distribution of Returns by Sector")
plt.suptitle("")  # remove the automatic "Boxplot grouped by" title
plt.xlabel("Sector")
plt.ylabel("YTD Return")
plt.xticks(rotation=45)

# Format y-axis as percentages
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f"{y:.0%}"))

plt.show()

```

### Financial Analysis
Tech is clearly a very large sector dominating the stock market which makes sense given it's prominence in the S&P 500. 

Finance although a smaller sector shows very good returns YTD as of the end of August 2025. Not sure what could be causing this but worth looking into. 

Retail is being dragged down quite heavily by Target and their ongoing boycott due to their stance on certain political issues. 

Auto holds steady with a smaller spread but returns are nothing to jump at.
